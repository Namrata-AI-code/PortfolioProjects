```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/weight-height.csv')
df.head()
```

```python
df.shape
```

```python
plt.hist(df.Height,bins=20,rwidth=0.8)
plt.xlabel('height')
plt.ylabel('count')
plt.show()
```

```python
plt.hist(df.Weight,bins=20,rwidth=0.8)
plt.xlabel('weight')
plt.ylabel('count')
plt.show()
```

```python
Q1 = df.Height.quantile(0.25)
Q3 = df.Height.quantile(0.75)
IQR = Q3 - Q1
IQR,Q1,Q3
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit,upper_limit
```

```python
height_outliers = (df.Height<lower_limit) | (df.Height>upper_limit)
height_outliers
```

```python
outliers_df = df[height_outliers]
outliers_df
```

```python
Q1w = df.Weight.quantile(0.25)
Q3w = df.Weight.quantile(0.75)
IQRw = Q3w - Q1w
lower_limit_w = Q1w - 1.5*IQRw
upper_limit_w = Q3w + 1.5*IQRw
lower_limit_w,upper_limit_w
```

```python
weight_outliers = (df.Weight<lower_limit_w) | (df.Weight>upper_limit_w)
w_outliers_df = df[weight_outliers]
w_outliers_df
```
