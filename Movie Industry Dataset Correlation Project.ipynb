```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) #Adjusts Size of plot we create
```

```python
df = pd.read_csv(r'C:\Users\Sougu\Downloads\movies.csv')
df.head()
```

```python
(df.isnull().sum()/len(df))*100
```

```python
df.shape
```

```python
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())*100
    print('{} : {}%'.format(col,pct_missing))
```

```python
#Dropping rows with missing values
df=df.dropna()
```

```python
#Checking data types of or columns
df.dtypes
```

```python
#To remove trailing zeroes after decimal point convert to int64
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df.head()
```

```python
#Create correct year column
df['year_correct'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head()
```

```python
df.sort_values(by = ['gross'],inplace=False,ascending=False)
df.head()
```

```python
#An option to look at all rows of data
pd.set_option('display.max_rows',None)
df.head()
```

```python
#Finding duplicates
#Shows only unique movie names - > df['name'].drop_duplicates().sort_values(inplace=False,ascending=False)
df=df.drop_duplicates()
df.head()
```

```python
#Finding correlation in the data

plt.scatter(data=df,x='budget',y='gross')
plt.title('Budget vs Gross Earnings')
plt.xlabel('budget for film')
plt.ylabel('gross earnings')
plt.show()
```

```python
#Plotting the same(Budget vs gross) using seaborn
sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
#Finding correlations in the data using correlation matrix
df.corr(method = 'pearson',numeric_only=True) #pearson, kendall, spearman
```

```python
#High correlation between budget and gross noted
correlation_matrix = df.corr(method = 'pearson',numeric_only=True)
#Visualizing using heatmap
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
#Converting non numerical data into numerical representations
df_numerized=df

for col_name in df_numerized.columns :
    if df_numerized[col_name].dtype == 'object':
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()        
        
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson',numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
df_numerized.corr()
```

```python
corr_matrix = df_numerized.corr()
corr_pairs = corr_matrix.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[sorted_pairs > 0.5]
high_corr
```
